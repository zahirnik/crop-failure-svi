# **Extract Climate Regions zip file**

```python
import zipfile
zip_path = '<DATA_ROOT>/US_ClimateRegions.zip'
extract_dir = '<DATA_ROOT>/US_ClimateRegions/'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
```

# **Spatial Join Climate Regions with Counties**

```python
import xarray as xr
import pandas as pd
import numpy as np
import geopandas as gpd
from matplotlib import pyplot as plt
import seaborn as sns
from glob import glob
```

```python
ClimReg_dir = '<DATA_ROOT>/US_ClimateRegions/US_ClimateRegions/'
gdf_clim = gpd.read_file(ClimReg_dir)
gdf_clim
```

```python
Counties_dir = '<DATA_ROOT>/SVI/Counties/'
gdf_counties = gpd.read_file(Counties_dir)
gdf_counties['geometry'] = gdf_counties['geometry'].centroid
gdf_counties
```

```python
joined_gdf = gpd.sjoin(gdf_counties, gdf_clim, how='left')
df_counties_clim = joined_gdf[['State', 'County', 'FIPS',	'ClimRegNam', 'ClimRegCod']]
df_counties_clim.to_csv('<DATA_ROOT>/US_ClimateRegions/Counties_ClimReg.csv')
```
