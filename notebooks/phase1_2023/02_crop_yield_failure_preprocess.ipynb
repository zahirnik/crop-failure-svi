```python
import pandas as pd

yield_file = '<DATA_ROOT>/Yield/crops_us_yield_area_harvested.csv'
df_yield = pd.read_csv(yield_file)
df_yield = df_yield.rename(columns={'statisticcat_desc_x':'statisticcat_desc',
                            'unit_desc_x':'unit_desc',
                            'domain_desc_x':'domain_desc'})

df_yield
```

```python
cond1 = df_yield['unit_desc']=='BU / ACRE'
df_yield[cond1].commodity_desc.unique()
```

```python
cond1 = df_yield['prodn_practice_desc']=='NON-IRRIGATED, CONTINUOUS CROP'
cond2 = df_yield['prodn_practice_desc']=='NON-IRRIGATED, FOLLOWING SUMMER FALLOW'
df_yield.loc[cond1,'prodn_practice_desc'] = 'NON-IRRIGATED'
df_yield.loc[cond2,'prodn_practice_desc'] = 'NON-IRRIGATED'

dict_bu2tons = {
    'WHEAT':0.027216, 'CORN':0.0254, 'SOYBEANS':0.027216,
    'BARLEY':0.021772, 'OATS':0.014515, 'SORGHUM':0.0254
    }

cond1 = df_yield['unit_desc']=='BU / ACRE'

for item in dict_bu2tons:
  print(item , dict_bu2tons.get(item))
  cond2 = df_yield['commodity_desc']== item
  bu2tons = dict_bu2tons.get(item)
  df_yield.loc[(cond1) & (cond2),'value_yield'] = df_yield.loc[(cond1) & (cond2),'value_yield'] * bu2tons

cond1 = df_yield['unit_desc'] == 'LB / ACRE'
df_yield.loc[cond1,'value_yield'] = df_yield.loc[cond1,'value_yield'] * 0.4535/1000
df_yield.unit_desc = 'TONS / ACRE'

df_yield['val_yield_area'] = df_yield['value_yield'] * df_yield['value_area']
grp_cols = ['commodity_desc' , 'prodn_practice_desc' , 'state_name' , 'county_name' , 'year']

df_yield_grp = df_yield.groupby(grp_cols).sum()
df_yield_grp['value_yield'] = df_yield_grp['val_yield_area'] / df_yield_grp['value_area']
df_yield_grp = df_yield_grp.drop(columns=['val_yield_area'])
df_yield_grp = df_yield_grp.reset_index()
df_yield_grp
```

```python
df_failure = pd.read_excel(files[-5])
df_failure_grp_first = df_failure.groupby('FIPS').first().reset_index()
df_failure_grp_first = df_failure_grp_first[['State','County','FIPS']]
df_failure_grp_first
```

```python
list_df_filure = []
for ffailure in files:
  year = int(ffailure.split('/')[-1][0:4])
  print(year)
  df_failure = pd.read_excel(ffailure)
  df_failure = df_failure.drop(columns=['Unnamed: 0'])

  group_cols_failure = ['FIPS' , 'Crop' ,  'Irrigation Practice']
  df_failure_grp_first = df_failure.groupby('FIPS').first().reset_index()
  df_failure_grp_first = df_failure_grp_first[['State','County','FIPS']]
  df_failure_grp = df_failure.groupby(group_cols_failure).sum()
  df_failure_grp = df_failure_grp.reset_index()
  df_failure_grp = pd.merge(df_failure_grp,df_failure_grp_first,on=['FIPS'])
  cond1 = df_failure_grp['Crop'] == 'ALFALFA'
  df_failure_grp.loc[(cond1),'Crop'] = 'HAY'

  if year < 2012:
    df_failure_with_all = df_failure_grp
  else:
    group_cols_sum_irig = ['FIPS' , 'Crop' ]
    df_failure_grp_irig = df_failure_grp.groupby(group_cols_sum_irig).sum()
    df_failure_grp_irig = df_failure_grp_irig.reset_index()
    df_failure_grp_irig = pd.merge(df_failure_grp_irig,df_failure_grp_first,on=['FIPS'])
    df_failure_grp_irig['Irrigation Practice'] = 'ALL'
    df_failure_with_all = pd.concat([df_failure_grp, df_failure_grp_irig],ignore_index=True)
  df_failure_with_all['year'] = year
  # break
  list_df_filure.append(df_failure_with_all)

df_failure_all = pd.concat(list_df_filure,ignore_index=True)
df_failure_all['County'] = df_failure_all['County'].str.capitalize()
```
