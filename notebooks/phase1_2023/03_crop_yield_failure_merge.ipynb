# **Yield Failure Merge**

```python
import pandas as pd

fpath_yield = '<DATA_ROOT>/Yield/crops_us_yield_agg_irig.csv'
fpath_failure = '<DATA_ROOT>/crop_failure_data/crops_failure_agg_irig.csv'

dfy = pd.read_csv(fpath_yield)
dff = pd.read_csv(fpath_failure)
```

```python
cond1 = dfy.commodity_desc == 'WHEAT'
cond2 = dfy.prodn_practice_desc == 'ALL PRODUCTION PRACTICES'
cond3 = dfy.state_name == 'WYOMING'
cond4 = dfy.county_name == 'WESTON'
dfy[cond1 & cond2 & cond3 & cond4][['year', 	'value_yield']].set_index('year').plot()
```

```python
# dfy.state_name   dff.State
# dfy.county_name dff.County
# dff.Irrigation Practice  I , N , ALL
# dfy.prodn_practice_desc 'ALL PRODUCTION PRACTICES' , 'IRRIGATED' , 'NON-IRRIGATED'
# year > 2008
```

```python
dfy['state_name'] = dfy['state_name'].str.capitalize()
dfy['county_name'] = dfy['county_name'].str.capitalize()

dff['State'] = dff['State'].str.capitalize()
dff['County'] = dff['County'].str.capitalize()

cond_A = dfy['prodn_practice_desc'] == 'ALL PRODUCTION PRACTICES'
cond_I = dfy['prodn_practice_desc'] == 'IRRIGATED'
cond_N = dfy['prodn_practice_desc'] == 'NON-IRRIGATED'

dfy.loc[cond_A , 'prodn_practice_desc'] = 'ALL'
dfy.loc[cond_I , 'prodn_practice_desc'] = 'I'
dfy.loc[cond_N , 'prodn_practice_desc'] = 'N'

dfy = dfy[dfy.year > 2008]

dfy
```

```python
dff = dff.drop(columns=['Unnamed: 0'])
dff
```

```python
rename_cols_dfy = {
    'commodity_desc' : 'Crop' , 'prodn_practice_desc' : 'Irrigation Practice' ,
    'state_name' : 'State' , 'county_name' : 'County'
}

dfy = dfy.rename(columns=rename_cols_dfy)
dfy
```

```python
join_cols = ['State',	'County', 'Crop', 'Irrigation Practice', 'year']
df_fail_yld = pd.merge(dff,dfy,on=join_cols,how='inner')
df_fail_yld#[df_fail_yld.value_yield>10]
```

```python
cond1 = df_fail_yld['Failed Acres'] > 0
df_fail_yld[cond1]
```

```python
import matplotlib.pyplot as plt

cond1 = df_fail_yld['County'] == 'Jackson'
cond2 = df_fail_yld['Crop'] == 'SOYBEANS'
cond3 = df_fail_yld['Irrigation Practice'] == 'ALL'
cond4 = df_fail_yld['State'] == 'Arkansas'
tmp = df_fail_yld[(cond1) & (cond2) & (cond3) & (cond4)]
tmp.plot(x='year', y='value_yield', kind='bar', title='Bar Plot')
tmp.plot(x='year', y='Failed Acres', kind='bar', title='Bar Plot')
```

```python
import matplotlib.pyplot as plt
tmp = tmp.sort_values(by='year')

x = tmp.year
yf = tmp['Failed Acres']
yy = tmp.value_yield
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.bar(x, yf )
plt.title('Failure Acres')
plt.subplot(1, 2, 2)
plt.bar(x,yy)
plt.title('Yield')
plt.tight_layout()
plt.show()
```
