# **Calcuate Crop Weighted SVI**

```python
# gdf2000,gdf2010,gdf2014,gdf2016,gdf2018,gdf2020
# df_area2008,df_area2010,df_area2020
# SVI_weightedCrop
```

```python
gdf2000 = gdf2000.set_index(['STCNTY', 	'FIPS' 	])
gdf2000 = gdf2000.applymap(lambda x: np.nan if x < 0 else x).reset_index()

gdf2010 = gdf2010.set_index(['STCNTY', 	'FIPS' 	])
gdf2010 = gdf2010.applymap(lambda x: np.nan if x < 0 else x).reset_index()

gdf2014 = gdf2014.set_index(['STCNTY', 	'FIPS' 	])
gdf2014 = gdf2014.applymap(lambda x: np.nan if x < 0 else x).reset_index()

gdf2016 = gdf2016.set_index(['STCNTY', 	'FIPS' 	])
gdf2016 = gdf2016.applymap(lambda x: np.nan if x < 0 else x).reset_index()

gdf2018 = gdf2018.set_index(['STCNTY', 	'FIPS' 	])
gdf2018 = gdf2018.applymap(lambda x: np.nan if x < 0 else x).reset_index()

gdf2020 = gdf2020.set_index(['STCNTY', 	'FIPS' 	])
gdf2020 = gdf2020.applymap(lambda x: np.nan if x < 0 else x).reset_index()
```

```python
gdf2000['FIPS'] = gdf2000['FIPS'].astype(int)

out_path = '<DATA_ROOT>/SVI_weightedCrop_modified/'
for year in [2008,2009]:
  df_area_y = df_area2008.loc[df_area2008.YEAR == year].drop(columns=['YEAR'])
  crops = ['HAY', 'BARLEY', 'OATS', 'CORN', 'COTTON', 'SORGHUM','SOYBEANS', 'WHEAT']
  for crop in crops:
    df_merge = pd.merge(gdf2000 , df_area_y[['FIPS',crop]],on=['FIPS']).drop(columns=['FIPS'])
    svi_idx = list(df_merge.columns)
    svi_idx.remove('STCNTY')
    svi_idx.remove(crop)
    df_merge[svi_idx] = df_merge[svi_idx].multiply(df_merge[crop], axis=0)
    df_group = df_merge.groupby('STCNTY').sum()
    df_group = df_group.divide(df_group[crop], axis=0)
    df_group = df_group.drop(columns=[crop])
    df_group['YEAR'] = year
    df_group.to_csv(out_path + 'SVI_' + crop + '_' + str(year) + '.csv')
```

# **Merge SVIs Crop by Crop**

```python
from glob import glob
in_path = '<DATA_ROOT>/SVI_weightedCrop_modified/'
out_path = '<DATA_ROOT>/SVI_weightedCrop_modified_Final/'
crops = ['HAY', 'BARLEY', 'OATS', 'CORN', 'COTTON', 'SORGHUM','SOYBEANS', 'WHEAT']
for crop in crops:
  files = sorted(glob(in_path + f'SVI_{crop}_*'))
  list_df = [pd.read_csv(fpath) for fpath in files]
  df = pd.concat(list_df, ignore_index=True)
  df.to_csv(out_path + 'SVI_' + crop + '.csv',index=False)
```
