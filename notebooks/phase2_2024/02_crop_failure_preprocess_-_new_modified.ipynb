# **Preprocess Crop Failure Data**

```python
import pandas as pd
from glob import glob

fpath = '<DATA_ROOT>/crop_failure_data/crop_failure_raw/*.xlsx'
files = sorted(glob(fpath))
files
```

```python
df0 = pd.read_excel(files[2], sheet_name='Sheet1')
df0 = df0.loc[:,['State Code' , 'State Name']]
df_st_codes = df0.groupby('State Name').first().reset_index()
df_st_codes
```

```python
df_2009 = pd.read_excel(files[0], sheet_name='Sheet1')
df_2009 = df_2009.drop(columns=['State Name'])
df_2009['FIPS'] = df_2009['State Code'] * 1000 + df_2009['County Code']
df_2009 = pd.merge(df_2009,df_st_codes, on=['State Code'], how='inner')
df_2009 = df_2009.drop(columns=['State Code','County Code','Crop Code','Crop Type Name'])

rename_cols_2009 = {
    'State Name' : 'State' , 'County Name' : 'County' , 'Crop Name' : 'Crop' , 'Crop Type' : 'Crop Type'
}
df_2009 = df_2009.rename(columns = rename_cols_2009)
cols = [
    'State', 'County','Crop','Crop Type', 'FIPS','Irrigation Practice',
    'Planted Acres','Prevented Acres','Failed Acres']
df_2009['Irrigation Practice'] = 'ALL'

state_name_mapping = pd.read_csv('<DATA_ROOT>/state_name_mapping.csv')
state_name_mapping = state_name_mapping.rename(columns={'State1':'State'})

df_2009 = pd.merge(df_2009, state_name_mapping, on=['State'], how='inner')
df_2009 = df_2009.drop(columns=['State'])
df_2009 = df_2009.rename(columns={'State2':'State'})
df_2009 = df_2009[cols]

crops = ['CORN', 'SOYBEANS', 'COTTON', 'WHEAT', 'BARLEY', 'OATS', 'ALFALFA', 'SORGHUM']
cond1 = df_2009['Crop'].str.contains('COTTON', case=False, na=False)
df_2009.loc[(cond1),'Crop'] = 'COTTON'

cond2 = df_2009['Crop'].str.contains('SORGHUM', case=False, na=False)
df_2009.loc[(cond2),'Crop'] = 'SORGHUM'

df_2009 = df_2009[df_2009['Crop'].isin(crops)]

cond3 = df_2009['Crop'] == 'ALFALFA'
df_2009.loc[(cond3),'Crop'] = 'HAY'

out_path = '<DATA_ROOT>/crop_failure_modified/crop_failure_final_20240321/'
fname = files[0].split('/')[-1]
print(fname)
df_2009.to_excel(out_path+fname)
```

# **Merge Crop Failure Data**

```python
import pandas as pd
from glob import glob
in_path = '<DATA_ROOT>/crop_failure_modified/crop_failure_final_20240321/*.xlsx'

files = sorted(glob(in_path))
files
```

```python
list_df = []

for f in files:
  year = int(f.split('/')[-1][0:4])
  dfi = pd.read_excel(f).copy()
  if year > 2011:
    dfi = dfi.rename(columns={'Planted and Failed Acres':'Planted Acres'})
  dfi = dfi.drop(columns=['Unnamed: 0'])
  dfi = dfi.groupby(['FIPS','Crop','Irrigation Practice']).agg({
      'State':'first','County':'first','Crop Type':'first',
      'Planted Acres':'sum','Prevented Acres':'sum','Failed Acres':'sum'}).reset_index()

  dfi_N = dfi[dfi['Irrigation Practice']=='N']
  dfi_I = dfi[dfi['Irrigation Practice']=='I']
  dfj = pd.merge(dfi_N,dfi_I[['FIPS','Crop','State','County','Planted Acres','Failed Acres']],on=['FIPS','Crop'],how='outer')#.fillna(0)
  nan_cols = ['Planted Acres_x','Failed Acres_x','Planted Acres_y','Failed Acres_y']
  dfj[nan_cols] = dfj[nan_cols].fillna(0)

  dfj['Planted Acres'] = dfj['Planted Acres_x'] + dfj['Planted Acres_y']
  dfj['Failed Acres'] = dfj['Failed Acres_x'] + dfj['Failed Acres_y']
  dfj['Irrigation Practice'] = 'ALL'

  dfj['State_x'] = dfj['State_x'].fillna(dfj['State_y'])
  dfj['County_x'] = dfj['County_x'].fillna(dfj['County_y'])
  dfj = dfj.rename(columns={'State_x':'State','County_x':'County'})

  cols = ['FIPS', 'Crop', 'Irrigation Practice', 'State', 'County', 'Planted Acres', 'Failed Acres']
  dfj = dfj[cols]
  dfi_N = dfi_N[cols]
  dfi_I = dfi_I[cols]

  dfi = pd.concat([dfi_I,dfi_N,dfj],ignore_index=True)
  dfi['year'] = year
  list_df.append(dfi)

df = pd.concat(list_df,ignore_index=True)
df
```

```python
df_fail = df.copy()
```

```python
cond1 = df_fail['Irrigation Practice'] == 'ALL'
cond2 = df_fail['Planted Acres'] > 0
percentiles = df[cond1 & cond2].groupby('Crop').quantile([0.05, 0.95])['Planted Acres']
percentiles = percentiles.reset_index()
percentiles = percentiles.pivot(index='Crop', columns='level_1', values='Planted Acres')
percentiles = percentiles.rename(columns = {0.05:'P05',0.95:'P95'}).reset_index()
percentiles
```

```python
cond1 = df_fail['Irrigation Practice'] == 'ALL'
df_fail_agg = df_fail[cond1].groupby(['FIPS','Crop','year']).sum().reset_index()
df_fail_agg = pd.merge(df_fail_agg,percentiles,on=['Crop'])

cond1 = df_fail_agg['Planted Acres'] > df_fail_agg['P05']
cond2 = df_fail_agg['Planted Acres'] < df_fail_agg['P95']
df_fail_agg = df_fail_agg[cond1 & cond2]
df_fail_agg = df_fail_agg.drop(columns=['Planted Acres','Failed Acres','P05','P95'])
df_fail_agg
df_fail = pd.merge(df_fail,df_fail_agg,on=['FIPS','Crop','year'])
```
